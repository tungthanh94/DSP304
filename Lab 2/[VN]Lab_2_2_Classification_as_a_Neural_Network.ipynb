## 3 - Kiến trúc chung của thuật toán học##

Đã đến lúc thiết kế một thuật toán đơn giản để phân biệt hình ảnh là cat (mèo) với non-cat.

Bạn sẽ xây dựng Hồi quy logistic, sử dụng tư duy Mạng nơ-ron. Hình sau giải thích lý do tại sao **Hồi quy Logistic thực sự là một mạng nơ-ron rất đơn giản!**

<img src="images/LogReg_kiank.png" style="width:650px;height:400px;">

**Biểu thức toán học của thuật toán**:

Cho một ví dụ $x^{(i)}$:
$$z^{(i)} = w^T x^{(i)} + b \tag{1}$$
$$\hat{y}^{(i)} = a^{(i)} = sigmoid(z^{(i)})\tag{2}$$ 
$$ \mathcal{L}(a^{(i)}, y^{(i)}) =  - y^{(i)}  \log(a^{(i)}) - (1-y^{(i)} )  \log(1-a^{(i)})\tag{3}$$

Cost (mất mát) được tính bằng cách tổng tất cả các ví dụ huấn luyện:
$$ J = \frac{1}{m} \sum_{i=1}^m \mathcal{L}(a^{(i)}, y^{(i)})\tag{6}$$

**Các bước chính**:
Trong tác vụ này, bạn sẽ thực hiện các bước sau: 
    - Khởi tạo các thông số của mô hình
    - Tìm hiểu các thông số cho mô hình bằng cách giảm thiểu mất mát  
    - Sử dụng các tham số đã học để đưa ra dự đoán (trên tập thử nghiệm)
    - Phân tích kết quả và kết luận

Nếu learning rate quá lớn, chúng ta có thể "vượt quá" giá trị tối ưu. Tương tự, nếu nó quá nhỏ, chúng ta sẽ cần rất nhiều lần lặp để hội tụ đến các giá trị tốt nhất. Đó là lý do tại sao điều quan trọng là phải sử dụng một learning rate được điều chỉnh tốt.
